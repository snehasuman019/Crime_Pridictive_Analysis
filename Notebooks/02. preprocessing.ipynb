```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv(("C:\\Users\\Sneha\\Desktop\\Crime Pridictive Analysis\\data\\Crime_Incidents_in_2025.csv"),encoding = "Latin 1")
print(df.head())
```

```python
df['REPORT_DAT'] = pd.to_datetime(df['REPORT_DAT'], errors='coerce')

df['Year'] = df['REPORT_DAT'].dt.year
df['Month'] = df['REPORT_DAT'].dt.month
df['Hour'] = df['REPORT_DAT'].dt.hour

df[['REPORT_DAT', 'Year', 'Month', 'Hour']].head()
# ML models cannot understand dates, but they understand:
# Year Month Hour
```

```python
# Handling missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
df['LATITUDE'].fillna(df['LATITUDE'].mean(), inplace=True)
df['LONGITUDE'].fillna(df['LONGITUDE'].mean(), inplace=True)

df.dropna(subset=['OFFENSE', 'METHOD', 'SHIFT'], inplace=True)
# Numerical → fill with mean
# Categorical → drop rows (small % loss is acceptable)
```

```python
le = LabelEncoder()

categorical_cols = ['OFFENSE', 'METHOD', 'SHIFT', 'DISTRICT']

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
scaler = StandardScaler()

numerical_cols = ['LATITUDE', 'LONGITUDE', 'Year', 'Month', 'Hour']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])

df[numerical_cols].head()
```

```python
df.to_csv("../data/crime_preprocessed.csv", index=False)
```
